#Criando DataFrame

```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas_Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs_Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas_Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)
display(df)
```

```python

# Adicionando colunas calculadas
df['Media_Horas_Trabalhadas'] = df['Horas_Trabalhadas'].mean()
df['Media_Bugs_Corrigidos'] = df['Bugs_Corrigidos'].mean()
df['Media_Tarefas_Concluídas'] = df['Tarefas_Concluídas'].mean()
df['Produtividade_Diaria'] = df['Tarefas_Concluídas'] / df['Horas_Trabalhadas']

# Imprimindo o DataFrame
display(df)

```

#Apresentando Gráficos:

## Horas trabalhadas

```python
# Calculando o total de horas trabalhadas e a média diária
total_horas_trabalhadas = df['Horas_Trabalhadas'].sum()
media_horas = round(df['Horas_Trabalhadas'].mean(), 2)

# Apresentando os cálculos
print('\033[1;34mTotal de Horas Trabalhadas: \033[0m', total_horas_trabalhadas)
print('\033[1;34mMédia Diária de Horas Trabalhadas: \033[0m', media_horas, '\n')

# Gráfico de Total de Horas Trabalhadas

plt.bar(df['Dia'], df['Horas_Trabalhadas'], color='blue')
plt.axhline(y=df['Horas_Trabalhadas'].mean(), color='red', linestyle='--', label='Média')
plt.title('Total de Horas Trabalhadas')
plt.legend()
```

##Bugs Corrigídos

```python
# Calculando o total de bugs corrigidos e sua média
total_bugs_corrigidos = df['Bugs_Corrigidos'].sum()
media_bugs = round(df['Media_Bugs_Corrigidos'].mean(), 2)

# Apresentando os cálculos
print('\033[1;34mTotal de Bugs corrigidos: \033[0m', total_bugs_corrigidos)
print('\033[1;34mMédia de Bugs corrigidos: \033[0m', media_bugs, '\n')


# Gráfico de Total de Bugs Corrigidos

plt.bar(df['Dia'], df['Bugs_Corrigidos'], color='orange')
plt.axhline(y=df['Bugs_Corrigidos'].mean(), color='red', linestyle='--', label='Média')
plt.title('Total de Bugs Corrigidos')
plt.legend()
```

## Tarefas Concluidas

```python
# Calculando o total de bugs corrigidos e sua média
total_tarefas = df['Tarefas_Concluídas'].sum()
media_tarefas = round(df['Media_Tarefas_Concluídas'].mean(), 2)

# Apresentando os cálculos
print('\033[1;34mTotal de tarefas concluídas: \033[0m', total_tarefas)
print('\033[1;34mMédia de tarefas concluídas: \033[0m', media_tarefas, '\n')


# Gráfico de Total de Tarefas Concluídas

plt.axhline(y=df['Bugs_Corrigidos'].mean(), color='red', linestyle='--', label='Média')
plt.bar(df['Dia'], df['Tarefas_Concluídas'], color='green')
plt.title('Total de Tarefas Concluídas')
```

##Produtividade diária

```python
# Gráfico de Produtividade Diária

plt.bar(df['Dia'], df['Produtividade_Diaria'], color='yellow')
plt.axhline(y=df['Produtividade_Diaria'].mean(), color='red', linestyle='--', label='Média')
plt.title('Produtividade Diária')
plt.legend()
```
